```python
import numpy as np
import pandas as pd
import seaborn as sns 
```

```python
df=pd.read_csv('air-quality-monitoring-sites (1).csv', delimiter=';')
df.head()
```

# PreProcessing


```python
import Preprocessing
nan = Preprocessing.NanAnalyzer(df)
nan.printNaNInfo()
```

```python
df= df.drop( columns= ['photopath','DateStart','DateEnd','Description','Colocated','Duplicate Triplicate','Grid_ID','Comments','Detailed location'])
```

```python
df.head()
```

```python
df= df.drop(df[(df['Distance from monitoring site to kerb'].isna() == True) & (df['Distance from receptor to kerb'].isna() == True)].index).reset_index(drop= True)
```

```python
df.head()
```

```python
nan = Preprocessing.NanAnalyzer(df)
nan.printNaNInfo()
```

```python
sample_height_mean=np.mean(df['Sample height'])
```

```python
df['Sample height'].fillna(value=sample_height_mean, inplace=True)
```

```python
df.head()
```

```python
nan = Preprocessing.NanAnalyzer(df)
nan.printNaNInfo()
```

```python
rows_with_nan = []
for index, row in df.iterrows():
    is_nan_series = row.isnull()
    if is_nan_series.any():
        rows_with_nan.append(index)

print(rows_with_nan)
```

```python
df.dropna(subset = ["Distance from monitoring site to kerb"], inplace=True)
df.dropna(subset = ["Distance from receptor to kerb"], inplace=True)
df.dropna(subset = ["Elevation"], inplace=True)
```

```python
nan = Preprocessing.NanAnalyzer(df)
nan.printNaNInfo()
```

```python
df.head()
```

```python
df= df.rename(columns={"SiteID":'site_id', "Easting":'easting', "Northing":'northing',"Current":'current',"Pollutants":'pollutants',"Instrument Type":'instrument_type',"Exposure":'exposure',"Distance from monitoring site to kerb":'Distance from monitoring site to kerb',"Distance from receptor to kerb":'Distance from receptor to kerb',"AQMA":'aqma',"Sample height":'sample_height',"Elevation":'elevation',"Location class":'location_class',"Location class LAQM":'location_class_laqm'})
```
